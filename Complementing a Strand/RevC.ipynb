```python
df = open("E:\\Biddle Lab\\Rosalind Project\\Complementing a Strand\\rosalind_revc.txt", "r")
```

```python
contents = df.read()
print(contents)
```

```python
x = ""
for i in reversed(contents):
    if i == "T":
        x += "A"
    if i == "A":
        x += "T"
    if i == "C":
        x += "G"
    if i == "G":
        x += "C"      
print(x)
```

```python
def revC(complement):
    revcontent = ""
    for i in complement[::-1]:
        if i == "T":
            revcontent += "A"
        if i == "A":
            revcontent += "T"
        if i == "C":
            revcontent += "G"
        if i == "G":
            revcontent += "C"      
    return(revcontent)
```

```python
print(contents)
rev = revC(contents)
print(rev)
```
