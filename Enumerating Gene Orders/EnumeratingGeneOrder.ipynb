```python
from parseRosalind import *
from itertools import permutations
import math
```

```python
def createPermString(file):
    permlen = parseFile(file)
    strperm = ""
    for i in range(1,permlen+1,1):
        stri = str(i)
        strperm += stri
    PermArray = [' '.join(p) for p in permutations(strperm)]
    print(len(PermArray))
    for elem in PermArray:
        print(elem)
```

```python
createPermString("rosalind_perm.txt")
```
